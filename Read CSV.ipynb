The loop (for y in x) directly prints each value, so only the values appear, not the indices.
Single Column: When you are dealing with only one column (like in your last example with just Calories), dropna() automatically applies to that column, so no subset is necessary.

```python
df = pd.read_csv("E:\\Download\\data.csv", usecols=["Calories"])
x = df["Calories"].dropna()
for y in x:
    print(y)
```

Two or More Columns: When you have multiple columns but want to drop rows based on all those columns, you can call dropna() without subset since it checks all specified columns by default.

```python
df = pd.read_csv("E:\\Download\\data.csv", usecols=["Calories", "Pulse"])
x = df.dropna()
print(x.to_string())
```

Decimal Values: The values in the "Calories" column (like 409.1, 479, and 282.4) cannot be converted to integers directly because integers cannot represent decimal numbers.

```python
df2 = pd.read_csv("E:\\Download\\data.csv", dtype={"Duration":"float"}) #convert int to float
df2
```

If you have a DataFrame with multiple columns but want to focus on dropping rows only based on NaNs in specific columns, then you would use subset. This allows you to ignore NaNs in other columns.

Using df2 makes the code more efficient and straightforward, while a new variable could unnecessarily complicate things.

```python
df2 = pd.read_csv("E:\\Download\\data.csv")
df2 = df2.dropna(subset=["Calories"])
df2["Calories"] = df2["Calories"].astype(int)
print(df2)
```
