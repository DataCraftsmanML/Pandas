Column Transformation Part 1

```python
import pandas as pd
```

```python
df = pd.read_excel("ESD.xlsx")
df.loc[(df["Bonus %"] == 0), "GetBonus % "] = "No  Bonus"
df.loc[(df["Bonus %"] > 0), "GetBonus % "] = "Bonus"
print(df)
```

```python
data = pd.read_excel("practice_1.xlsx")
print(data, "\n")
data["Full Name"] = data["First Name"].str.capitalize() + " " + data["Last Name"].str.capitalize()
print(data)
```

Column Transformation Part 2

```python
data = pd.read_excel("Practice_1.xlsx")
data["Bonus"] = (data["Salary"]/100)*20
data["Bonus"] = data["Bonus"].astype(int)
data
```

```python
data["Bonus Received"] = (data["Bonus"] / data["Salary"]) * 100
data["Bonus Received"] = data["Bonus Received"].apply(lambda x: f"{x:.0f}%")
data
```

```python
data["Bonus"] = data["Bonus"].apply(lambda x: x * 2)
data
```

```python
data = data.rename(columns={'Bonus':'Bonus ₹', 'Bonus Received':'Bonus Received %'})
data
```

```python
data1 = {"Month":["january", "february", "march", "april"]}
x = pd.DataFrame(data1)

def extract (Value):
    return Value[0:3]
    
x["Short Month"] = x["Month"].map(extract)
x
```
